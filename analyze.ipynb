```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def draw_landmarks(img, lmks, color=(255,0,0)):
    img = np.ascontiguousarray(img)
    for a in lmks:
        cv2.circle(img,(int(round(a[0])), int(round(a[1]))), 1, color, -1, lineType=cv2.LINE_AA)

    return img

lmsk = np.loadtxt("/home/vuthede/Face2FaceRHO/test_case/driving/FLAME/landmarkde.txt")
lmsk = (lmsk+1)/2 * 512
img = np.zeros((512,512,3))

img = draw_landmarks(img, lmsk)
plt.imshow(img)    
```
